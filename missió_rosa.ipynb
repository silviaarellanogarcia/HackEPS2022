```python
img_mal = cv2.imread('gdrive/MyDrive/HackEPS2022/data/rosa/A3/to/c1r1e0n2-28.png')

color = ('b','g','r')

for i, c in enumerate(color):
    hist = cv2.calcHist([img_mal], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])

plt.show()

cv2.destroyAllWindows()
```
